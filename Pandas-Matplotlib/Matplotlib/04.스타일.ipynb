# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] # Mac
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, y, linewidth=10) # 라인 두께 설정
```

## 마커(marker)

```python
plt.plot(x, y, marker='o') # 표시
```

```python
plt.plot(x, y, marker='o', linestyle='None') #선 없애기
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor = 'red') # 바깥색 설정
```

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor = 'red', markerfacecolor='yellow') # 안에 색 설정
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':') # 점선
```

```python
plt.plot(x, y, linestyle='--') # 두꺼운 점선
```

```python
plt.plot(x, y, linestyle='-.') # 선과 점
```

```python
plt.plot(x, y, linestyle='-') # 실선
```

## 색깔

```python
plt.plot(x, y, color='pink') # 선 색 설정
```

```python
plt.plot(x, y, color='#ff0000') # rgb로 색 설정
```

```python
plt.plot(x, y, color='b') # 파란색
```

```python
plt.plot(x, y, color='g') # 초록색
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:') # 파란색 v 점선
```

```python
plt.plot(x, y, 'go') #초록색 o 선없음
```

```python
plt.plot(x, y, color ='g', marker='o', linestyle='None')
```

## 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms='10', mec='blue', ls = ':') # 줄여쓰기
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms='10', alpha=0.7) # alpha : 투명도(0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5)) # (가로, 세로)
plt.plot(x,y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10, 5), dpi=200) # dots per inch, 확대
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor='yellow') # 배경색 설정
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x, y)
```
